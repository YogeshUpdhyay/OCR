```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
img = cv2.imread('../sample_cropped.jpg', 0)
```

```python
img.dtype
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
# binaristaion of the image
ret_bin, img_bin = cv2.threshold(img,160,255,cv2.THRESH_BINARY)
plt.imshow(img_bin)
```

```python
# otsu binarization
img_bin_otsu = cv2.threshold(img, 0, 255,  cv2.THRESH_OTSU)[1]
plt.imshow(img_bin_otsu)
```

```python
med_img = cv2.inRange(img, 160, 200)
plt.imshow(med_img)
```

```python
cv2.imwrite("../sample_binary.jpg", img_bin_otsu)
```
