```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Show image

```python
image = cv2.imread('sample.jpg')
plt.imshow(image)

```

## Rotate the image

```python
image_rotate = cv2.rotate(image, cv2.cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(image_rotate)
```

## Size

```python
print(image_rotate.shape)
print(image_rotate.size)
print(image_rotate.dtype)
```

## Rescaling

```python
scale_percent = 200 
width = int(image_rotate.shape[1] * scale_percent / 100)
height = int(image_rotate.shape[0] * scale_percent / 100)
dsize = (width, height)
image_resize = cv2.resize(image_rotate, dsize)
plt.imshow(image_resize)
print(image_resize.size)
```

## Background Removal (Canny Edge Detection)

```python
edges = cv2.Canny(image_resize,100,200)
cv2.imwrite('cannyedge.jpg',edges)
canny_edge = cv2.imread('cannyedge.jpg')
plt.imshow(canny_edge)
print(canny_edge.shape)
```

## Binarization

```python
im_gray = cv2.cvtColor(image_resize, cv2.COLOR_BGR2GRAY)

th, im_gray_th_otsu = cv2.threshold(im_gray, 128, 192, cv2.THRESH_OTSU)

print(th)
# 117.0

cv2.imwrite('opencv_th_otsu.jpg', im_gray_th_otsu)
binarization = cv2.imread('opencv_th_otsu.jpg')
plt.imshow(binarization)
print(binarization.shape)
```

## Adaptive Thresholding

```python
img_th = cv2.cvtColor(image_resize, cv2.COLOR_BGR2GRAY)
thresh2 = cv2.adaptiveThreshold(img_th, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY, 199, 5)
cv2.imwrite('adap_th.jpg', thresh2)
adap_th = cv2.imread('adap_th.jpg')
plt.imshow(adap_th)
print(adap_th.shape)
```
